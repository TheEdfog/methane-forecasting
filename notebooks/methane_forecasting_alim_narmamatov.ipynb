This notebook is a short walkthrough. The reusable implementation lives in the `methane_forecasting` package.

```python
import pandas as pd

from methane_forecasting.data import ensure_dataset
from methane_forecasting.features import build_supervised_frame
from methane_forecasting.model import train_forecaster

# Uses data/filtered_data_frame.csv when present. Otherwise downloads the original file.
dataset = ensure_dataset()
data = pd.read_csv(dataset)
data.head()
```

## Feature construction

At one reading per second, 900 rows correspond to a 15 minute forecast horizon. Rolling features use historical values and the train-test split remains chronological.

```python
features, target = build_supervised_frame(data, horizon_steps=900)
features.shape, target.shape
```

```python
model, report = train_forecaster(features, target)
report.to_dict()
```

The report includes the model MAE and the persistence MAE. Comparing both values is more informative than reporting the model metric alone.
